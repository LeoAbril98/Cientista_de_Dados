```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd = pd.read_csv('Estrutura_SINASC_para_CD.csv')
```

```python
pd.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=pd)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=pd)
sns.countplot(ax = axes[1, 0], x='SEXO', data=pd)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=pd)
```

```python
pd['GESTACAO'].value_counts().sort_index().plot.bar()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5',y = 'CONSULTAS', data=pd)
```
